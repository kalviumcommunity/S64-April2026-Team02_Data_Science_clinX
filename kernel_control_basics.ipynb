## Interrupt vs Restart

Interrupt is used to stop a currently running cell without clearing notebook memory.

Restart is used to completely reset the kernel and clear all variables from memory.

Interrupt is useful for infinite loops or slow cells.
Restart is safer when the notebook state becomes inconsistent.
